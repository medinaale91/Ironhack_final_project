```python
# Creamos una funcion para obtener los datos de la web, pasandole como parametro el número
# de pagina(nos interesan las 15 primeras páginas)
def get_marathon_records(page):

    url = "https://www.worldathletics.org/records/all-time-toplists/road-running/marathon/outdoor/men/senior"
    params = {
              "regionType":"world",
              "drop":"regular",
              "fiftyPercentRule":"regular",
              "page":page,
              "bestResultsOnly":"false",
              "firstDay":"1899-12-31",
              "lastDay":"2020-12-09"
             }
    data = requests.get(url, params=params)
    soup = BeautifulSoup(data.text)
    data_scraped = []
    rows = soup.select(".records-table tbody tr")
    for row in rows:
        athlete = []
        for content in row.select("td"):
            if len(content.text) > 0:
                athlete.append(content.text.strip())
        data_scraped.append(athlete)
    return data_scraped
```

```python
class MarathonRecord:
    def __init__(self, rank, mark, competitor, dob, nat, pos, venue, date, value_score):
        self.rank = rank
        self.mark = mark
        self.competitor = competitor
        self.dob = dob
        self.nat = nat
        self.pos = pos
        self.venue = venue
        self.date = date
        self.value_score = value_score
    
    def __str__(self):
        return f"{self.rank} {self.mark} {self.competitor} {self.dob} {self.nat} {self.pos} {self.venue} {self.date} {self.value_score}"

def get_marathon_records(page):
    # esta función devuelve una lista de MarathonRecords para la página concreta que se nos pasa
    # para hacer scraping
    url = "https://www.worldathletics.org/records/all-time-toplists/road-running/marathon/outdoor/men/senior"
    params = {"page":page}
    data = requests.get(url, params=params)
    soup = BeautifulSoup(data.text)
    marathon_records = []
    rows = soup.select(".records-table tbody tr")
    for row in rows:
        tds = row.select("td")
        marathon_record = MarathonRecord(
            rank = tds[0].text.strip(),
            mark = tds[1].text.strip(),
            competitor = tds[2].text.strip(),
            dob = tds[3].text.strip(),
            nat = tds[4].text.strip(),
            pos = tds[5].text.strip(),
            venue = tds[7].text.strip(),
            date = tds[8].text.strip(),
            value_score = tds[9].text.strip()
        )
        marathon_records.append(marathon_record)
    return marathon_records
```
