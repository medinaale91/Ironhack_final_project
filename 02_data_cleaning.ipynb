```python
import pandas as pd
```

```python
df_top_7100_men = pd.read_csv("./output/data_7100_top_men.csv")
```

```python
df_top_7100_men
```

```python
df_top_7100_men.drop(columns=[ "Unnamed: 0", "0","8"], inplace=True) # Eliminamos columnas de escasa relevancia.
```

```python
df_top_7100_men.columns = ["Mark","Competitor","DOB","Nat","Pos","Venue","Date"]
df_top_7100_men
```

```python
df_top_7100_men.rename_axis('Rank', inplace=True)
df_top_7100_men
```

```python
df_top_7100_men.info() #observamos valores nulos en la columna DOB y Pos.
```

```python
df_top_7100_men.dropna(inplace=True) #los eliminamos al ser irrelevantes para nuestro ejercicio.
df_top_7100_men.info()
```

```python
df_top_7100_men.Competitor.value_counts()
```

```python
# Convert DOB and Date columns to datetime so we can operate with them and get two 
# more columns: "Age at race" and "Current Age".

df_top_7100_men['Date'] = pd.to_datetime(df_top_7100_men.Date)
df_top_7100_men['DOB'] = pd.to_datetime(df_top_7100_men.DOB)

df_top_7100_men.info()
```

```python
# AÃ±adir nueva Columna: Age when the race happened. Aportaria valor para averiguar la edad
# ideal, la que mÃ¡s estÃ¡ repetida en este dataset.

import datetime

def calculate_age(since, until):
    return until.year - since.year - ((until.month, until.day) < (since.month, since.day))

# Nos creamos una lista vacia en la que aÃ±adiremos las edades cuando se produjo la carrera.
# Mediante la funcion iterrows, nos devuelve el indice y la fila (el indice no 
# nos interesa realmente), calculamos la edad con la funcion calculate_age, pasandole 
# como parametros row["DOB"] y row["Date"].

men_age_at_race = []
for index, row in df_top_7100_men.iterrows():
    men_age_at_race.append(calculate_age(row["DOB"], row["Date"]))
    
men_age_at_race
```

```python
# insertamos la columna "Age at race" al final del dataset.
df_top_7100_men.insert(7, "Age at race", men_age_at_race)
df_top_7100_men
```

```python
#aÃ±adir nueva Columna: Current Age

men_current_age = []
for index, row in df_top_7100_men.iterrows():
    men_current_age.append(calculate_age(row["DOB"], datetime.datetime.now()))
    
men_current_age
```

```python
# insertamos la columna "Current Age" al final del dataset.
df_top_7100_men.insert(8, "Current age", men_current_age) 
df_top_7100_men
```

```python
# Creamos una nueva columna "Year of the race" cogiendo el aÃ±o de la columna "Date". La insertamos al final del dataset. 
df_top_7100_men['Year of the race']=df_top_7100_men.Date.dt.year
df_top_7100_men
```

```python
#Function to convert running time in seconds

import datetime

def timestamp_to_sec(time):
    try:
        return (datetime.datetime.strptime(time, "%H:%M:%S.%f") - datetime.datetime(1900, 1, 1)).total_seconds()
    except:
        return (datetime.datetime.strptime(time, "%H:%M:%S") - datetime.datetime(1900, 1, 1)).total_seconds()
```

```python
df_top_7100_men["TimeSec"] = df_top_7100_men["Mark"].apply(timestamp_to_sec)
df_top_7100_men
```

```python
df_top_7100_men.to_csv("output/data_7100_top_men_cleaned.csv")
```

```python
# Creamos otra tabla con los 1500 mejores registros, el resto no los cogemos.
df_top_1500_men=df_top_7100_men[:1500]
df_top_1500_men
```

```python
from dotenv import load_dotenv
import os
import requests
import re
import time
```

```python
load_dotenv()
def temperature(city, date):
    weather_token = os.getenv("weather_token")
    url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/weatherdata/history?aggregateHours=24&combinationMethod=aggregate&startDateTime={date}T00%3A00%3A00&endDateTime={date}T00%3A00%3A00&maxStations=-1&maxDistance=-1&contentType=json&unitGroup=metric&locationMode=single&key={weather_token}&dataElements=default&locations={city}"
    data = requests.get(url)
    weather = data.json()
    if "location" in weather:
        return weather["location"]["values"][0]["temp"]
    else:
        return None
```

```python
#iterrows() Iterate over DataFrame rows as (index, Series) pairs.
temps = []

for index, values in df_top_1500_men.iterrows():
    city = re.sub(r'\s\([^)]*\)', '', values["Venue"]) # nuestra API no acepta el cÃ³digo de paÃ­s en el nombre, la limpiamos mediante regex
    date = values["Date"].strftime("%Y-%m-%d") #Values de date es un objeto de tipo datetime, por lo que necesitamos convertirlo a string en el formato que espera la API (YYYY-MM-DD).
    temp = temperature(city, date)
    print("Temperature for ", city, " at ", date, " was ", temp)
    temps.append(temp)
    time.sleep(1)

temps
```

```python
temps
```

```python
df_top_1500_men.insert(10, "Temp", temps)
df_top_1500_men
```

```python
df_top_1500_men.to_csv("output/data_1500_top_men_cleaned.csv")
```
