```python
import pandas as pd
```

```python
df = pd.read_csv("./output/data.csv")
```

```python
df
```

```python
df.drop(columns=[ "Unnamed: 0", "0","8"], inplace=True) # Eliminamos columnas de escasa relevancia.
```

```python
df.columns = ["Mark","Competitor","DOB","Nat","Pos","Venue","Date"]
df
```

```python
df.rename_axis('Rank', inplace=True)
df
```

```python
df.info() #observamos 9 valores nulos en la columna DOB.
```

```python
df.dropna(inplace=True) #los eliminamos al ser irrelevantes para nuestro ejercicio.
df.info()
```

```python
df.Competitor.value_counts()
```

```python
# Convert DOB and Date columns to datetime so we can operate with them and get two 
# more columns: "Age at race" and "Current Age".

df['Date'] = pd.to_datetime(df.Date)
df['DOB'] = pd.to_datetime(df.DOB)

df.info()
```

```python
# Añadir nueva Columna: Age when the race happened. Aportaria valor para averiguar la edad
# ideal, la que más esta repetida en este dataset.

import datetime

def calculate_age(since, until):
    return until.year - since.year - ((until.month, until.day) < (since.month, since.day))

# Nos creamos una lista vacia en la que añadiremos las edades cuando se produjo la carrera.
# Mediante la funcion iterrows, nos devuelve el indice y la fila (el indice no 
# nos interesa realmente), calculamos la edad con la funcion calculate_age, pasandole 
# como parametros row["DOB"] y row["Date"].

age_at_race = []
for index, row in df.iterrows():
    age_at_race.append(calculate_age(row["DOB"], row["Date"]))
    
age_at_race
```

```python
df.insert(7, "Age at race", age_at_race) # insertamos la columna al final del dataset
df
```

```python
#añadir nueva Columna: Current Age

current_age = []
for index, row in df.iterrows():
    current_age.append(calculate_age(row["DOB"], datetime.datetime.now()))
    
current_age
```

```python
df.insert(8, "Current age", current_age) # insertamos la columna al final del dataset
df
```

```python
df.to_csv("output/data_cleaned.csv")
```
